5. What is mutability in Python?
Ans 5-
Mutability in Python means that they can be changed after creation. Examples: Lists, Dictionaries, Sets, etc.

6. Why are list mutable, but tuples are immutable?
Ans 6-
A Python list is essentially a dynamic array under the hood. It stores references to objects and can grow or shrink in size. A tuple is stored as a fixed size sequence of object references. Once created, Python doesnt allow modifications.

7. What is the difference between "==" and "is" operators in Python?
Ans 7-
"==" is equality operator and it checks if the values of two objects are the same. "is" is identity operator and it checks if two variables point to the same object in the memory.

8. What are logical operators in Python?
Ans 8-
In Python, logical operators are used to combine or invert conditions. There are 3 main logical operators 'and', 'or' and 'not'.

9. What is type casting in Python?
Ans 9-
 Type casting is the process of cahnging a variable from one data type to another.

10. What is the difference between implicit and explicit type casting?
Ans 10-
Implicit type casting is the kind where conversion is done automatically by Python while in Explicit type casting the data type is manually converted by the programmer.

11. What is the purpose of conditional statements in Python?
Ans 11-
Conditional statements controls the flow of programme by executing different code blocks depending on wether a condition is true or false.

12. How does the elif statement work?
Ans 12-
Elif stands for "else-if". It lets us check mutiple conditions one after the other. Only the first condition that evaluates to True gets executed. Once a True branch is found, Python skips the rest.

13. What is the difference between for and while loops?
Ans 13-
For loop is used when we know how many times we want to repeat something. While loop is used when we dont know beforehand how many times to loop but we want to continue untill a condition becomes false.

14. Describe a scenario where a while loop is more suitable than a for loop?
Ans 14-
A while loop is suitable than a for loop where we dont know how many attempts the user will need. Example of such scenario is Game Loops (run untill the player quits).
